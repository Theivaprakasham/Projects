```python
df = pd.read_csv("/content/UCI HAR Dataset/train/X_train.txt", delim_whitespace=True, header=None);df.head()
```

```python
#Dataset Preparation

X_train = pd.read_csv('/content/UCI HAR Dataset/train/X_train.txt',delim_whitespace=True,header=None).to_numpy()
y_train = pd.read_csv('/content/UCI HAR Dataset/train/y_train.txt',delim_whitespace=True,header=None).to_numpy().reshape(-1)
X_test = pd.read_csv('/content/UCI HAR Dataset/test/X_test.txt',delim_whitespace=True,header=None).to_numpy()
y_test = pd.read_csv('/content/UCI HAR Dataset/test/y_test.txt',delim_whitespace=True,header=None).to_numpy().reshape(-1)
```

**Support Vector Machine (SVM)**




```python
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#SVM with RBF Kernal
classifier = svm.SVC(gamma=0.001, kernel='rbf')
classifier.fit(X_train, y_train)
svm_predicted = classifier.predict(X_test)

print(classification_report(y_test, svm_predicted))
print("Confusion Matrix\n", ConfusionMatrix(y_test, svm_predicted))
```

```python
# SVM WIth Linear Kernal

classifier = svm.SVC(gamma=0.001, kernel='linear')
classifier.fit(X_train, y_train)
svm_predicted = classifier.predict(X_test)

print(classification_report(y_test, svm_predicted))
print("Confusion Matrix\n", ConfusionMatrix(y_test, svm_predicted))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
  
# defining hyperparameter range
param_grid = {'gamma': [1, 0.01, 0.001],
              'kernel': ['rbf', 'linear']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)


# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)

grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions))
print("Confusion Matrix\n", ConfusionMatrix(y_test, grid_predictions))
```

SVM Classifier was used to predict six Human activity classes.

RBF Kernal and Linear Kernals were used intially to test. We also perfomed a grid search to arrive at the best Hyperparammeter Gamma.

**RBF Kernal**: Accuracy -> **93%**

**Linear Kernal**: Accuracy -> **96%**


**From the results of the Grd Search, we observe that SVM with Linear Kernal with Gamma=1 was able to give a maximum accuracy of 96%.**

**Decision Tree Classifier**

```python
from sklearn import tree

Dtree = tree.DecisionTreeClassifier()
Dtree.fit(X_train, y_train)

Dtree_predicted = Dtree.predict(X_test)

print(classification_report(y_test, Dtree_predicted))
print("Confusion Matrix\n", ConfusionMatrix(y_test, Dtree_predicted))

#tree.plot_tree(Dtree) 
```

```python
dec_tree = tree.DecisionTreeClassifier()

# defining hyperparameter range
param_grid = {'max_depth': [1, 50, 100, 150, 250]} 
  
grid = GridSearchCV(dec_tree, param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)


# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)

grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions))
print("Confusion Matrix\n", ConfusionMatrix(y_test, grid_predictions))
```

Decision Tree Classifier was used to predict six Human activity classes.

We also perfomed a grid search to arrive at the best Hyperparameter Max_depth.

Accuracy -> **87%**

From the results of GridSearch, we observe that Decision Tree Classifier was able to give a maximum accuracy of **87%** with Hyper-Paramaeter **max_depth: 100**.

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_predicted = rf.predict(X_test)

print(classification_report(y_test, rf_predicted))
print("Confusion Matrix\n", ConfusionMatrix(y_test, rf_predicted))
```

```python
# defining hyperparameter range
param_grid = { 'max_depth': [5, 50, 100], 'n_estimators': [5, 50, 100, 150]}

grid = GridSearchCV(rf, param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)


# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)

grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions))
print("Confusion Matrix\n", ConfusionMatrix(y_test, grid_predictions))
```

Random Forest Classifier was used to predict six Human activity classes.

We also perfomed a grid search to arrive at the best Hyperparammeter Max_depth and n-estimator(Maximum number of trees).

Accuracy -> **92%**

From the results of GridSearch, we observe that Random Forest Classifier gave a maximum accracy of **92%** with **'max_depth': 100**, **'n_estimators': 150**
