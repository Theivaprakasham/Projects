```python
def plotsir(t, S, I, R):
  f, ax = plt.subplots(1,1,figsize=(10,4))
  ax.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
  ax.plot(t, I, 'y', alpha=0.7, linewidth=2, label='Infected')
  ax.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recovered')

  ax.set_xlabel('Time (days)')

  ax.yaxis.set_tick_params(length=0)
  ax.xaxis.set_tick_params(length=0)
  #ax.grid(b=True, which='major', c='w', lw=2, ls='-')
  legend = ax.legend()
  legend.get_frame().set_alpha(0.5)
  for spine in ('top', 'right', 'bottom', 'left'):
      ax.spines[spine].set_visible(False)
  plt.show();
```

```python
def deriv(y, t, N, beta, gamma):
  S, I, R = y
  dSdt = -beta * S * I / N
  dIdt = beta * S * I / N - gamma * I
  dRdt = gamma * I
  return dSdt, dIdt, dRdt
```

```python
N = 3000000 # Toal Population
beta = 1/6  # infected person infects 5 persion in 15 days with 50% probability ->pc' = Transmission Rate
D = 15 # infections lasts fifteen days
gamma = 1.0 / D # Recovery Rate when infections lasts D days

S0, I0, R0 = 2999900, 100, 0  # initial conditions: one recovered, two infected, rest susceptible

t = np.linspace(0, 200, 200) # Grid of time points (in days)
y0 = S0, I0, R0 # Initial conditions vector

# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

plotsir(t, S, I, R)
print("\nNo of persons infected: ", max(I))
```

**f. If the disease is deadly and some of the infected individuals die, how can this be incorporated in the model ? Death rate can be taken as 2 % (Out of 100 people getting infected, 2 individuals may die). What changes will happen to spread of the epidemic?**

```python
def deriv_death(y, t, N, beta, gamma, k):
  S, I, R, D = y
  dSdt = -beta * S * I / N
  dIdt = beta * S * I / N - gamma * I - (k*I)
  dRdt = gamma * I
  dDdt = k * I
  return dSdt, dIdt, dRdt, dDdt
```

```python
def plotsird(t, S, I, R, D):
  f, ax = plt.subplots(1,1,figsize=(10,4))
  ax.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
  ax.plot(t, I, 'y', alpha=0.7, linewidth=2, label='Infected')
  ax.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recovered')
  ax.plot(t, D, 'r', alpha=0.7, linewidth=2, label='Dead')

  ax.set_xlabel('Time (days)')

  ax.yaxis.set_tick_params(length=0)
  ax.xaxis.set_tick_params(length=0)
  #ax.grid(b=True, which='major', c='w', lw=2, ls='-')
  legend = ax.legend()
  legend.get_frame().set_alpha(0.5)
  for spine in ('top', 'right', 'bottom', 'left'):
      ax.spines[spine].set_visible(False)
  plt.show();
```
